# 모집단분포와 확률표본

## 모집단과 모수

* 모집단은 어떤 통계적 실험의 대상이 되는 모든 대상물
* 모집단을 구성하는 데이터가 이루는 확률분포가 모집단분포(population distribution)

* 모수(parameter)는 모집단의 특성을 나타내는 수치로 모평균, 모분산, 모표준편차, 모비율, 모상관관계 등이 있다. 
* 즉, 모집단을 구성하는 데이터를 설명하는 수치 값
* 모수가 중요한 이유는 실제로 알고 싶어 하는 (추정하고 싶어 하는) 것은 '모집단 데이터'가 아니라 '모집단 특성을 설명하는 모수'이기 때문이다. 
* 그리고 대부분의 데이터에서 모집단 분포는 완전하게 알려진 것이 없다.  모집단이 추상적이기 때문에 완전한 데이터를 얻을 수 없으며, 얻을 수 있더라도 경제적, 공간적, 시간적 제약으로 거의 불가능하다

## 모집단과 확률표본
* 모집단의 모수(특성)를 알 수 있는 방법은 표본을 추출해서 모집단의 모수를 추정하는 것이다.
* 그렇다면 표본을 추출하는 방식은 모든 데이터가 뽑힐 가능성(확률)을 동등하게 부여하며, 객관적(실험자 의도가 부여되지 않고)으로 무작위(랜덤) 추출한 표본이어야 한다. 
* 이러한 표본을 확률표본(random sample)이라고 한다

## 회귀식과 회귀분석
* 회귀분석은 변수 간의 함수 관계를 분석하느 ㄴ방법 중 하나로 독립변수와 종속변수 간의 1차 선형적 관계를 도출하여 독립변수가 종속변수에 미치는 영향 혹은 예측 정도를 분석하는 방법이다
	* 회귀(regression) : '평균으로의 회귀'를 의미, 두 변수 관계가 일반화된 선형 관계의 평균으로 돌아감을 의미
	* 선형성(linearity): 두 변수 관계를 직선(1차 방정식 : Y = aX + b) 하나의 형태로 설명할 수 있다.

## 잔차와 오차

* 회귀식은 모집단을 대표하는 직선을 의미하는데, 모집단을 가진 경우는 거의 없기 때문에 모집단을 대표할 수 있는 표본집단의 회귀식을 추론한다. 여기에서 나온 개념이 잔차(residual)와 오차(error)이다
* 잔차와 오차의 개념을 혼동해서 사용할 때가 있는데, 
	*	잔차  = (표본집단의 회귀식에서 예측된 값) - (실제 관측 값)
	*	오차 = (모집단의 회귀식에서 예측된 값) - (실제 관측 값)
